```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.image import imread
import plotly
import random
import seaborn as sns
import joblib
```

```python
RESULTS_PATH = '../input/shopee-competition-results/'
```

```python
tfidf_resnet_union_df = pd.read_csv(RESULTS_PATH+'tfidf_resnet_union_df.csv')
tfidf_resnet_union_df
```

```python
nfnet_sbert_union_df = pd.read_csv(RESULTS_PATH+'nfnet_sbert_union_df.csv')
nfnet_sbert_union_df
```

```python
tfidf_resnet_ensemble_df = pd.read_csv(RESULTS_PATH+'tfidf_resnet_ensemble_df.csv')
tfidf_resnet_ensemble_df
```

```python
nfnet_sbert_ensemble_df = pd.read_csv(RESULTS_PATH+'nfnet_sbert_ensemble_df.csv')
nfnet_sbert_ensemble_df
```

```python
text_image_union_df = pd.read_csv(RESULTS_PATH+'text_image_union_df.csv')
text_image_union_df
```

```python
image_text_results = [[0 for i in range(3)] for j in range(5)]

image_text_results[0][0] = round(tfidf_resnet_union_df['f1'][0],6)
image_text_results[0][1] = round(tfidf_resnet_union_df['recall'][0],6)
image_text_results[0][2] = round(tfidf_resnet_union_df['precision'][0],6)

image_text_results[1][0] = round(nfnet_sbert_union_df['f1'][0],6)
image_text_results[1][1] = round(nfnet_sbert_union_df['recall'][0],6)
image_text_results[1][2] = round(nfnet_sbert_union_df['precision'][0],6)

image_text_results[2][0] = round(tfidf_resnet_ensemble_df['f1'][0],6)
image_text_results[2][1] = round(tfidf_resnet_ensemble_df['recall'][0],6)
image_text_results[2][2] = round(tfidf_resnet_ensemble_df['precision'][0],6)

image_text_results[3][0] = round(nfnet_sbert_ensemble_df['f1'][0],6)
image_text_results[3][1] = round(nfnet_sbert_ensemble_df['recall'][0],6)
image_text_results[3][2] = round(nfnet_sbert_ensemble_df['precision'][0],6)

image_text_results[4][0] = round(text_image_union_df['f1'][0],6)
image_text_results[4][1] = round(text_image_union_df['recall'][0],6)
image_text_results[4][2] = round(text_image_union_df['precision'][0],6)
```

```python
image_text_results_df = pd.DataFrame(image_text_results,columns=['f1','recall','precision'],index=['tfidf resnet union','nfnet sbert union','tfidf resnet ensemble','nfnet sbert ensemble','text image union'])
image_text_results_df
```

```python
image_text_results_df.to_csv('image_text_results_df.csv')
```

```python
ax = image_text_results_df.plot(marker='o',ylabel='score',rot=70)
# plt.legend(loc='center left')
fig = ax.get_figure()
fig.savefig('../doc/figures/image_text_results_df.pdf',pad_inches=0,bbox_inches='tight')
```
